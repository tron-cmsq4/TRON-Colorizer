```python
import os

path_to_data = os.path.abspath('George_W_Bush')
image_names = os.listdir(path_to_data)
raw_data = []

for image in image_names:
    raw_data.append(imread(path_to_data+'/'+image))
```

```python
compressed_gs_data = []

for image in raw_data:
    compressed_image, compression = compressSVD(grayscale(image))
    compressed_gs_data.append(compressed_image)
```

```python
for i in range(len(raw_data)):
    raw_data[i] = raw_data[i].flatten().astype('int')
    compressed_gs_data[i] = compressed_gs_data[i].flatten().astype('int')
```

```python
#zero pad images
total_length_raw = len(max(raw_data,key=len))
total_length_compressed = len(max(compressed_gs_data,key=len))

for i, image in enumerate(raw_data):
    pad_amount = total_length_raw - len(image)
    image = np.pad(image, (0,pad_amount), 'constant', constant_values=(0,0))
    
for i, image in enumerate(compressed_gs_data):
    pad_amount = total_length_compressed - len(image)
    image = np.pad(image, (0,pad_amount), 'constant', constant_values=(0,0))
```

```python
from sklearn.model_selection import train_test_split

gs_train, gs_test, raw_train, raw_test = train_test_split(raw_data, compressed_gs_data,
                                                         test_size=0.25, random_state=13)
```
