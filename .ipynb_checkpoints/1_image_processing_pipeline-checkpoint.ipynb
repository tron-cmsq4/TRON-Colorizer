```python
## compress png images

import cv2
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import time
from matplotlib.image import imread
from PIL import Image


## Nick's code

def grayscale(img_orig):
    img_gs = img_orig[:, :, 0] * 0.21 + img_orig[:, :, 1] * 0.72 + img_orig[:, :, 2] * 0.07
    return img_gs



## takes an array of input values (RGB) and then reads the image as an array and compresses that
## Code adapted from Frank Cleary

def compressSVDTest(img_gs):      ## img_gs = grayscale image array
    plt.figure(figsize=(9, 6))
    plt.imshow(img_gs)

    U, sigma, V = np.linalg.svd(img_gs)
    
    reconstimg = np.matrix(U[:, :1]) * np.diag(sigma[:1]) * np.matrix(V[:1, :])
    plt.imshow(reconstimg, cmap='gray')
    
    ## find optimal values for i
    for i in range(5, 51, 5): # start, stop, step
        reconstimg = np.matrix(U[:, :i]) * np.diag(sigma[:i]) * np.matrix(V[:i, :])
        plt.imshow(reconstimg, cmap='gray')
        title = "n = %s" % i
        plt.title(title)
        plt.show()
    
    
```

```python
my_img = imread("https://4.bp.blogspot.com/-SDEf6RuSD7Q/V_o_r9pUjyI/AAAAAAAACSw/9fP4fbzKc7gL5XDyIo5-eU1ZU3V29MheACLcB/s1600/Screen%2BShot%2B2016-01-19%2Bat%2B15.16.36.png")
plt.imshow(grayscale(my_img), interpolation = 'nearest', cmap = 'gray')
plt.show()
```

```python
my_gs_img = grayscale(my_img)
compressSVDTest(my_gs_img) ## WORKS!!!!!!
```

```python
def compressSVD(img_gs):      ## img_gs = grayscale image array

    U, sigma, V = np.linalg.svd(img_gs)
     
    reconstimg = np.matrix(U[:, :50]) * np.diag(sigma[:50]) * np.matrix(V[:50, :])
    reconstimg = np.array(reconstimg)
    plt.imshow(reconstimg, cmap='gray')
    compression_ratio = 100.0*(50*img_gs.shape[0] + img_gs.shape[1] + 50) / (img_gs.shape[0] * img_gs.shape[1])
    
    return reconstimg, compression_ratio
```
